```python
import pandas as pd
import matplotlib.pyplot as plt
```

```python
PATH = r'C:\Users\CarlosABG\Documents\IMP\inflacao 2017'
df = pd.read_excel(PATH + '/dados.xlsx')
```

```python
# USER SETTINGS

LEFT = -50
RIGHT = 40
BOTTOM = -1
TOP = 110

TITLE = "Brasil: Inflação de 400 produtos diferentes, 2017"
SUBTITLE = "Em porcento, variação acumulada em 12 meses"

X_LABEL = "Variação acumulada no preço em 12 meses (%)"
Y_LABEL = "Número de produtos com essa variação"

SOURCE = 'Índice de Preços ao Consumidor Amplo, IBGE.'
```

```python
# CODE

plt.style.use('fivethirtyeight')

# Declare figure
fig, ax = plt.subplots(figsize=(12,7))

# Plot histogram
plt.hist('inflation', bins=30, color='brown', alpha=1, data=df)

# Configure axes
ax.yaxis.grid(which="major", color='grey', linewidth=3)
ax.xaxis.grid(which="major", linewidth=0)
ax.tick_params(axis='both', which='major', labelsize=15)
ax.set_xlim(left = LEFT, right = RIGHT)
ax.set_ylim(bottom = BOTTOM, top = TOP)

# Set line at zero
plt.axvline(0, color='grey', linewidth=3)
plt.axhline(0, color='grey', linewidth=3)

# Set titles
ax.text(x = LEFT-7, y = TOP+8, s = TITLE,
               fontsize = 26, weight = 'bold')
ax.text(x = LEFT-7, y = TOP+2, 
               s = SUBTITLE,
              fontsize = 19)

# Set axis labels
ax.set_xlabel(X_LABEL, fontsize=19)
ax.set_ylabel(Y_LABEL, fontsize=19)

# Set source:

ax.text(x =  LEFT-7, y = BOTTOM-17,
    s = 'Fonte: ' + SOURCE,
    fontsize = 14)

# Set label below
ax.text(x = LEFT-7, y = BOTTOM-24,
    s = 'Instituto Mercado Popular' + ' '*130 + 'www.mercadopopular.org',
    fontsize = 14, color = '#f0f0f0', backgroundcolor = 'grey')
plt.show()
```
